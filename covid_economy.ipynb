```python
import pandas as pd
import numpy as np
```

```python
covid_data = pd.read_csv('raw_data.csv')
```

```python
#Filter location to be Afghanistan
```

```python
#Create a table of total cases and find daily new cases
df= pd.DataFrame(covid_data, columns = ['location', 'date', 'total_cases', 'population'])
print(df.head())
```

```python
df=df.sort_values(by=['location','date'])
```

```python
daily_new_cases=[]
for country in df['location'].unique(): 
    x=df[df['location']==country]['total_cases'].diff()
    daily_new_cases.append(x.values)
#df['daily_new_cases']=daily_new_cases
```
